```python
hemi_url='https://astrogeology.usgs.gov/search/results?q=hemisphere+enhanced&k1=target&v1=Mars'
```

```python
browser.visit(hemi_url)
```

```python
hemis=['Cerberus', 'Schiaparelli', 'Syrtis', 'Valles']
```

```python
hemi_request=requests.get('https://astrogeology.usgs.gov/search/map/Mars/Viking/cerberus_enhanced')
soup=bs(hemi_request.text, 'lxml')
soup.find('img', class_='wide-image')['src']
soup.find('h2', class_='title').text
```

```python
hemisphere_image_urls=[]
for x in hemis:
    browser.click_link_by_partial_text(x)
    html=browser.html
    soup=bs(html, 'lxml')
    img_url=soup.find('img', class_='wide-image')['src']
    title=soup.find('h2', class_='title').text
    dict={
        'title':title,
        'img_url':(f'https://astrogeology.usgs.gov{img_url}')
    }
    hemisphere_image_urls.append(dict)
    browser.visit(hemi_url)
    
```
